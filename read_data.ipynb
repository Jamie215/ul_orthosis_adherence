```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
# Load the csv files
emg_rest_df = pd.read_csv("EMG_DATA_REST.CSV", encoding="utf-8")
emg_flex_df = pd.read_csv("EMG_DATA_FLEX_1MIN.CSV", encoding="utf-8")
emg_activity_df = pd.read_csv("EMG_DATA_ACTIVITY.CSV", encoding="utf-8")

# Convert the "Time(ms)" column to numeric for plotting
emg_rest_df["Time(ms)"] = pd.to_numeric(emg_rest_df["Time(ms)"])
emg_flex_df["Time(ms)"] = pd.to_numeric(emg_flex_df["Time(ms)"])
emg_activity_df["Time(ms)"] = pd.to_numeric(emg_activity_df["Time(ms)"])

# Plot the EMG signal data
print(len(emg_rest_df))
print(len(emg_flex_df))
print(len(emg_activity_df))

fig, ax = plt.subplots(3)
ax[0].plot(emg_rest_df["EMG Signal"][400:])
ax[1].plot(emg_flex_df["EMG Signal"][400:])
ax[2].plot(emg_activity_df["EMG Signal"][400:])

ax[0].set_xlabel("Time (ms)")
ax[0].set_ylabel("EMG Signal")
ax[1].set_xlabel("Time (ms)")
ax[1].set_ylabel("EMG Signal")
ax[2].set_xlabel("Time (ms)")
ax[2].set_ylabel("EMG Signal")
plt.show()
```
